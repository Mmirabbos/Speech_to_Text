# Train

```python
def load_dataset():
    return pd.read_csv("data.csv")

df = load_dataset()

def load_spectrogram(path):
    image = tf.io.read_file(path)
    image = tf.image.decode_png(image, channels=3)
    image = tf.image.resize(image, [230, 230])
    image = image / 255.0

    return image

spectrogram_path_folder = os.listdir("spectrograms")
spectrogram_os_path = r"C:\Users\mirabbos\Desktop\qwerty\spectrograms"

spectrogram_paths = [os.path.join(spectrogram_os_path, filename) for filename in spectrogram_path_folder]

spectrogram_data = tf.data.Dataset.from_tensor_slices(spectrogram_paths)

spectrogram_data = spectrogram_data.map(load_spectrogram)

text_labels = df['transcribe'].tolist()
label_dataset = tf.data.Dataset.from_tensor_slices(text_labels)

dataset = tf.data.Dataset.zip((spectrogram_data, label_dataset))

train_size = int(0.7 * len(dataset))
val_size = int(0.15 * len(dataset))
test_size = len(dataset) - train_size - val_size

train_dataset = dataset.take(train_size)
val_dataset = dataset.skip(train_size).take(val_size)
test_dataset = dataset.skip(train_size + val_size)

print("Размер обучающего набора:", len(train_dataset))
print("Размер валидационного набора:", len(val_dataset))
print("Размер тестового набора:", len(test_dataset))
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(230, 230, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(3)
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(train_dataset, validation_data=val_dataset, epochs=10)

val_loss, val_accuracy = model.evaluate(val_dataset)
print("Валидационная потеря:", val_loss)
print("Валидационная точность:", val_accuracy)

test_loss, test_accuracy = model.evaluate(test_dataset)
print("Тестовая потеря:", test_loss)
print("Тестовая точность:", test_accuracy)
```
